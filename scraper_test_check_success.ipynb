```python
import random
from time import sleep

import undetected_chromedriver as uc

from scrape.utils import filter_html
from scrape.utils import get_listings
from scrape.utils import get_page_count
from scrape.utils import make_page_urls
from scrape.utils import get_all_listings

from scrape.constants import FREGUESIA_URL
```

```python
# 1. open driver load first page

FREGUESIA_URL = "https://www.idealista.pt/arrendar-casas/lisboa/"
driver = uc.Chrome()
driver.get(FREGUESIA_URL)
sleep(3)
raw_html = driver.page_source
```

```python
# 2. get page count
# make urls

pages = get_page_count(raw_html)
urls = make_page_urls(pages, FREGUESIA_URL)
```

```python
for url in urls:
    driver.get(url)
    raw_html = driver.page_source
    listings_list, validation_status = get_listings(raw_html)
    print(f"url:{url} scraped, status{validation_status}")
```
